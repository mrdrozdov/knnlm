```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
knn = [
(2, 0.2916),
(4, 0.2907),
(8, 0.2890),
(16, 0.2864),
(32, 0.2840),
(64, 0.2821),
]
optimal = [
(2, 0.7623),
(4, 0.7574),
(8, 0.7459),
(16, 0.7213),
(32, 0.6750),
(64, 0.5907),
]
optimal_v2 = [
(2, 0.6895),
(4, 0.6197),
(8, 0.5488),
(16, 0.4787),
(32, 0.3992),
(64, 0.3212),
]
```

```python
plt.figure()
ks, probs = zip(*knn)
plt.plot(ks, probs, label='KNN-LM', color='black', linewidth=1)
ks, probs = zip(*optimal)
plt.plot(ks, probs, label='Optimal', color='black', linewidth=4)
ks, probs = zip(*optimal_v2)
plt.plot(ks, probs, label='Optimal w/o Neg.', color='black', linestyle='dashed')
plt.xscale('log', basex=2)
plt.legend(loc='best')
plt.ylabel('Average Next Word Probability')
plt.xlabel('Number of Retrieved Items ($K$)')
plt.savefig('upper_bound.pdf', dpi=300)
plt.savefig('upper_bound.png', dpi=300)
plt.show()
```
